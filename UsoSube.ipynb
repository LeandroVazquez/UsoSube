```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Campos de los dataset

dia_transporte | Fecha ISO-8601 (date) | día de transporte informado |
---------------|-----------------------|-----------------------------|
amba | Texto (string) | SI/NO |
motivo_atsf | Texto (string) | Motivo del Atributo Social Federal |
genero | Texto (string) | F/M/nulo (esto último para cuando se trate de tarjetas no registradas) |
tipo_transporte | Texto (string) | colectivo, tren, subte, lancha o total (total para AMBA o total para INTERIOR). Tener en cuenta que el total no va a ser igual a la suma por medio, ya que una misma tarjeta pudo haber viajado en más d |
cant_trj | Número entero (integer) | cantidad de tarjetas únicas con alguna transacción de uso en el día. Las transacciones de uso son Uso, Checkin, Checkout, venta de boleto, y sus eventuales reversas. |  
dato_preliminar | Texto (string) | SI/NO 

```python
#Cargar el primer documento que contiene los datos
data2020=pd.read_csv("dat-ab-usuarios-2020.csv")
data2020
```

```python
#Cargar el segundo documento que contiene los datos
data2021=pd.read_csv("dat-ab-usuarios-2021.csv")
data2021
```

```python
#Cargar el tercer documento que contiene los datos
data2022=pd.read_csv("dat-ab-usuarios-2022.csv")
data2022
```

```python
#Concateno los tres documentos en uno
tablaConcatenada= pd.concat([data2020,data2021,data2022], axis=0)
tablaConcatenada
```

```python
#Separo la fecha en Año, Mes y Día.
tablaGlobal=tablaConcatenada
tablaGlobal["Tiempo"]=tablaGlobal["DIA_TRANSPORTE"].astype(str)
tablaGlobal["Año"]=tablaGlobal["Tiempo"].str[0:4]
tablaGlobal["Mes"]=tablaGlobal["Tiempo"].str[5:7]
tablaGlobal["Día"]=tablaGlobal["Tiempo"].str[8:10]
tablaGlobal
```

```python
#Función que permite aplicar un filtro que selecciona solo los totales
def filtrarTotales():
    filtroTotales=tablaGlobal["TIPO_TRANSPORTE"]=="TOTAL"
    return filtroTotales
```

```python
#Funcion que permite crear el índice de años
def crearIndice():
    indiceAnios=["2020","2021","2022"]
    return indiceAnios
```

```python
tablaAmba=tablaGlobal[filtrarTotales()]
filtroAmbaNo=tablaAmba["AMBA"]=="NO"
auxiliarAmbaNo=tablaAmba[filtroAmbaNo]
tablaAmbaNo=auxiliarAmbaNo.groupby("Año")["CANT_TRJ"].mean()

filtroAmbaSi=tablaAmba["AMBA"]=="SI"
auxiliarAmbaSi=tablaAmba[filtroAmbaSi]
tablaAmbaSi=auxiliarAmbaSi.groupby("Año")["CANT_TRJ"].mean()

tablaGeneral=pd.DataFrame({"No":tablaAmbaNo,"Si":tablaAmbaSi},index=crearIndice())
##tablaGeneral.plot(kind="bar",grid=True)
ax= tablaGeneral.plot(kind="bar",grid=True)
plt.title("Promedio de Tarjetas Usadas, separadas por si lo fueron en AMBA o no")
plt.xlabel("Año")
plt.ylabel("Promedio, expresada en 1e8")
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

```python
##tablaTransporte=tablaGlobal[filtrarTotales()]
filtroColectivo=tablaGlobal["TIPO_TRANSPORTE"]=="COLECTIVO"
auxiliarColectivo=tablaGlobal[filtroColectivo]
tablaColectivo=auxiliarColectivo.groupby("Año")["CANT_TRJ"].mean()

filtroLancha=tablaGlobal["TIPO_TRANSPORTE"]=="LANCHA"
auxiliarLancha=tablaGlobal[filtroLancha]
tablaLancha=auxiliarLancha.groupby("Año")["CANT_TRJ"].mean()

filtroSubte=tablaGlobal["TIPO_TRANSPORTE"]=="SUBTE"
auxiliarSubte=tablaGlobal[filtroSubte]
tablaSubte=auxiliarSubte.groupby("Año")["CANT_TRJ"].mean()

filtroTren=tablaGlobal["TIPO_TRANSPORTE"]=="TREN"
auxiliarTren=tablaGlobal[filtroTren]
tablaTren=auxiliarTren.groupby("Año")["CANT_TRJ"].mean()

tablaTransportes=pd.DataFrame({"Colectivo":tablaColectivo,"Lancha":tablaLancha,"Subte":tablaSubte,"Tren":tablaTren},index=crearIndice())
ax = tablaTransportes.plot.bar(grid=True)
plt.title("Promedio de tarjetas usadas separadas por el tipo de transporte"")
plt.xlabel("Año")
plt.ylabel("Promedio")

for container in ax.containers:
    ax.bar_label(container)
    
plt.show()
```

```python
tablaMotivo=tablaGlobal[filtrarTotales()]

filtroAUH=tablaMotivo["MOTIVO_ATSF"]=="AUH"
auxiliarAUH=tablaMotivo[filtroAUH]
tablaAUH=auxiliarAUH.groupby("Año")["CANT_TRJ"].mean()

filtroJubilacion=tablaMotivo["MOTIVO_ATSF"]=="JUBILACION"
auxiliarJubilacion=tablaMotivo[filtroJubilacion]
tablaJubilacion=auxiliarJubilacion.groupby("Año")["CANT_TRJ"].mean()

filtroMonotributo=tablaMotivo["MOTIVO_ATSF"]=="MONOTRIBUTO SOCIAL"
auxiliarMonotributo=tablaMotivo[filtroMonotributo]
tablaMonotributo=auxiliarMonotributo.groupby("Año")["CANT_TRJ"].mean()

filtroOtro=tablaMotivo["MOTIVO_ATSF"]=="OTRO"
auxiliarOtro=tablaMotivo[filtroOtro]
tablaOtro=auxiliarOtro.groupby("Año")["CANT_TRJ"].mean()

filtroPension=tablaMotivo["MOTIVO_ATSF"]=="PENSION"
auxiliarPension=tablaMotivo[filtroPension]
tablaPension=auxiliarPension.groupby("Año")["CANT_TRJ"].mean()

filtroProgresar=tablaMotivo["MOTIVO_ATSF"]=="PROGRESAR"
auxiliarProgresar=tablaMotivo[filtroProgresar]
tablaProgresar=auxiliarProgresar.groupby("Año")["CANT_TRJ"].mean()

filtroDomestico=tablaMotivo["MOTIVO_ATSF"]=="PERSONAL DEL TRABAJO DOMÉSTICO"
auxiliarDomestico=tablaMotivo[filtroDomestico]
tablaDomestico=auxiliarDomestico.groupby("Año")["CANT_TRJ"].mean()

tablaMotivos=pd.DataFrame({"AUH":tablaAUH,"Jubilación":tablaJubilacion,"Monotributo Social":tablaMonotributo,"Otro":tablaOtro,"Pensión":tablaPension,"Progresar":tablaProgresar,"Personal del Trabajo Doméstico":tablaDomestico},index=crearIndice())
ax=tablaMotivos.plot.bar(grid=True)

for container in ax.containers:
    ax.bar_label(container)

plt.legend(bbox_to_anchor=(1, 1))
plt.show()
```
